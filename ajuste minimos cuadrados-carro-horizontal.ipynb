```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("bmh")
```

```python
datos= np.genfromtxt("carro-plano-01.txt")
t= datos [:,0]
x= datos [:,1]
```

```python
plt.scatter(t,x)
plt.grid(True)
plt.xlabel("$t \\rm [s]$")
plt.ylabel("$x$[m]")
plt.xlim(0,3)
plt.ylim(0,1.3)
```

```python
def g(x0,v):
    tt = np.linspace(0,3,100)
    xx = x0 + v*tt
    plt.figure()
    plt.plot(tt,xx)
    plt.grid(True)
    plt.xlabel("$t \\rm [s]$")
    plt.ylabel("$x$[m]")
    plt.xlim(0,3)
    plt.ylim(0,1.3)
    chi2=np.sum(((x0+v*t-x)**2))
    plt.title("$\\chi^2 =$"+ str(chi2))
```

```python
from ipywidgets import interact
```

```python
interact(g, x0 = (-10,10,0.01), v=(-10,10,0.01))
```
