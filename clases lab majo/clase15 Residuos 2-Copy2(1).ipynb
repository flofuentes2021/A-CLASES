```python
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
```

# Ejemplo 1 : modelo lineal + error simulado 

```python
N = 70 
x = np.linspace(0,10,N)
y = x + 0.2*(np.random.rand(N)-0.5)
np.savetxt('ejemplo-s11a.txt', np.stack((x,y), axis=-1#limpia columnas), fmt="%.3f" #redondea decimales)
```

```python
b1,b0 = np.polyfit(x,y,1)
xx = np.linspace(-1,11,10000)
plt.scatter(x,y)
plt.plot(xx,b0+b1*xx, color='red', linewidth=2)
```

```python
#Coeficiente de determinación 
St = np.sum((y-np.mean(y))**2)
Sr = np.sum((y-b0-b1*x)**2)
r2 = (St - Sr)/St
print(r2)
```

```python
#Coeficiente de posición y pendiente
print(b0,b1)
```

```python
#Residuos
ep = y - (b0+b1*x)
#grafico de los residuos 
plt.scatter(x,ep)
plt.grid()
```
