```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-5,5,10)
def f(x):
    return np.sin(x)/x
```

```python
plt.style.use('classic')
x = np.linspace(-20,20,100)
plt.plot(x,f(x), label='primera funcin')
plt.plot(x,f(x)/2, label='segunda funcin')
plt.grid(True)
plt.xlabel('$x$')
plt.ylabel('$f(x)=\\sin(x)/x$')
plt.legend(loc='best')
plt.savefig('Segundogrficos22.pdf')
```

```python
plt.style.use('classic')
N=2
x = np.linspace(-20,20,N)
plt.plot(x,f(x), marker='.')
plt.grid(True)
plt.xlabel('$x$')
plt.ylabel('$f(x)=\\sin(x)/x$')
```

```python
plt.style.use('classic')
N=1000
x = np.linspace(-20,20,N)
plt.plot(x,f(x), marker='.')
plt.grid(True)
plt.xlabel('$x$')
plt.ylabel('$f(x)=\\sin(x)/x$')
```

```python
plt.style.use('classic')
for N in range(2,100,2):
    x = np.linspace(-20,20,N)
    plt.figure()
    #se reinicia y crea graficos nuevos, o sea no crea graficos sonre el anterior
    #esto lo hace hasta el grafico con N=98 ya que es un ciclo
    plt.plot(x,f(x), marker='.')
    plt.grid(True)
    plt.xlabel('$x$')
    plt.ylabel('$f(x)=\\sin(x)/x$')
    plt.title('Graficos con N='+str(N))
    #plt.savefig('grafico-N-'+str(N)+'.pdf')
```
