Ahora usaremos matplotlib para confeccionar un gr√°fico de la funci√≥n
\begin{equation}
f(x)=\frac{sin(x)}{x}
\end{equation}
en el intervalo $x\in[-5,5]$

```python
x = np.linspace(-5,5,10)
y = np.sin(x)/x
print(x)
print(y)
plt.plot(x,y, marker='o')
```

```python
#insert cell ab en insert
def f(x):
    return np.sin(x)/x
#definimos una funcion f(x)
```

```python
x = np.linspace(-20,20,100)
plt.plot(x,f(x), label='primera funci√≥n')
plt.plot(x,f(x)/2, label='segunda funci√≥n')
#los label describen lo que es cada curva con su repectivo color
plt.grid(True)
plt.xlabel('$x$')
plt.ylabel('$f(x)=\\sin(x)/x$')
plt.legend(loc='best')
plt.savefig('Segundogr√°fico.pdf')
#hace funcionar los label
#loc indica la posici√≥n de la simbolog√≠a de label, en este caso best es la que considera la mejor posicion
#aqupi tambi√©n se pueden poner numeros que determina en que esquina ponerlo
#dar muchos puntos da la iluci√≥n de que la curva es suave
#en MATPLOLOTLIB.STYLES se pueden ver los estilos de gr√°ficos que tiene
```
