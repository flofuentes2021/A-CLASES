```python
n = np.array([0,1,2,3,4,5])
```

```python
plt.figure(figsize=(10,4))
#Define una figura nueva de 10x4
plt.subplot(1,4,1)
#Arreglo de 1x4 figuras, 1er gr√°fico
plt.scatter(n, n**2)
#Scatter eran los gr√°ficos s√≥lo de puntos
plt.subplot(1,4,2)
#Arreglo de 1x4 figuras, 2do gr√°fico
plt.step(n, n**2, lw=2)
#Step es un gr√°fico de escal√≥n
plt.subplot(1,4,3)
#Arreglo de 1x4 figuras, 3er gr√°fico
plt.bar(n, n**2, align="center", width=0.5, alpha=0.5)
#Bar son gr√°ficos de barras
plt.subplot(1,4,4)
#Arreglo de 1x4 figuras, 4to gr√°fico
plt.fill_between(x, x**2, x**3, color="green", alpha=0.5)
#fill between es un gr√°fico relleno entre ambas funciones x**2  y x***3
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,3,1)
plt.scatter(n, n**2)
plt.subplot(1,3,2)
plt.step(n, n**2, lw=2)
plt.subplot(1,3,3)
plt.bar(n, n**2 )
#(datos eje x, datos eje y o altura,with=0.8 por defecto qque es el ancho, align=center o edge
#que es la alineaci√≥n de la barra, edge lo alinea a la izquierda, alpha=x controla la transparencia, 
#botton=x las barras comienzan desde x
plt.grid(True)
```

```python
alpha = 0.7
phi = 2*np.pi*0.5

def flux_qubit_potential(x,y):
    return 2 + alpha - 2 * np.cos(2*np.pi*x)*np.cos(2*np.pi*y) - alpha*np.cos(phi-2*x)
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
X,Y = np.meshgrid(x, y)
Z = flux_qubit_potential(X, Y).T
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
#es para ejecutar figuras en 3d
```

```python
ax = plt.subplot(1, 1, 1, projection='3d')
ax.plot_surface(X, Y, Z, rstride=1, cstride=1,  linewidth=0, antialiased=False)
```
