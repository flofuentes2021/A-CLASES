Los residuos muestran una oscilaci√≥n, que podr√≠amos modelar con la funci√≥n de tipo: $\alpha \sin{(\beta x + \gamma)}$

Esto significa que podemos rehacer el ajuste  dde nuestros datos (x,y) usando la siguiente ecuaci√≥n modificada: 
\begin{equation}
\bar{f}(x) = a_1 x + a_0 + \alpha \sin{(\beta x + \gamma)}
\end{equation}

```python
def f(x,a1,a0,a,b,g):
    return a1*x + a0 + b*np.sin(b*x + g)
```

```python
from scipy.optimize import curve_fit
```

```python
a1,a0,a,b,g = curve_fit(f,x,y)[0]
print('a1 = ', a1)
print('a0 = ', a0)
print('alpha = ', a)
print('beta = ', b)
print('gamma = ', g)
```

```python
xx = np.linspace(0,10,1000)
plt.plot(xx,f(xx,a1,a0,a,b,g), color= 'red')
plt.scatter(x,y)
plt.grid(True)
plt.xlabel('$x$')
plt.ylabel('$y$')
```
