```python
alturas = np.array([1.65,1.84,1.53,1.87,1.65,1.75,1.58,1.59,1.77,1.83,1.68,1.80,1.59,1.70,1.79,1.66,1.81,1.53])
```

```python
len(alturas)
```

```python
s = np.std(alturas,ddof=1)
p = np.mean(alturas)
plt.scatter(np.arange(len(alturas)),alturas)
plt.ylim(1,2)
#limites en el eje y
plt.hlines(p,0,17)
#una linea en y=prom que vaya en x desde 0 hasta el 17
plt.hlines(p+s,0,17,color='red')
plt.hlines(p-s,0,17,color='red')
#lineas de las desviaciones estandar, el error del promedio
plt.grid(True)
```

Tenemos que el promedio se calcula de la siguiente manera:
\begin{equation}
\bar{x} := \frac{1}{N}\sum_{i=1}^N x_i,
\end{equation}
Con $x$ con datos desde $i$ hasta $N$ valores

```python
prom = np.sum(alturas)/len(alturas)
print(prom)
```

```python
p = np.mean(alturas)
#promedio de los datos
```

Por otro lado tenemos la Desviación estándar que se calula de la siguiente manera:
\begin{equation}
s := \sqrt{\frac{1}{N-1}\sum_{i=1}^N (x_i- \bar{x})^2}
\end{equation}
Y la varianza:
\begin{equation}
s^2 := \frac{1}{N-1}\sum_{i=1}^N (x_i- \bar{x})^2
\end{equation}

```python
s = np.sqrt(np.sum((alturas- p)**2)/(len(alturas)-1))
s
```
