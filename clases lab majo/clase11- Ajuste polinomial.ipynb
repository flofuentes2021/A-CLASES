```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos = np.genfromtxt('C01.txt')
```

```python
t = datos[:,0]
x = datos[:,1]
```

```python
plt.scatter(t,x)
plt.grid(True)
plt.xlabel('$t$ [s]')
plt.ylabel('$x$ [m]')
```

```python
def g(a0,a1):
    plt.scatter(t,x)
    plt.grid(True)
    plt.xlabel('$t$ [s]')
    plt.ylabel('$x$ [m]')
    xx = np.linspace(np.min(t),np.max(t),100)
    yy = a0+a1*xx
    plt.plot(xx,yy)
```

```python
a1 = (np.mean(t*x)-np.mean(t)*np.mean(x))/(np.mean(t**2)-(np.mean(x**2)))
```

```python
a1
```

```python
a0 = np.mean(x)-a1*np.mean(t)
```

```python
a0
```

```python
g(a0,a1)
```

```python
a1,a0 = np.polyfit(t,x,1)
```

```python
g(a0,a1)
```
