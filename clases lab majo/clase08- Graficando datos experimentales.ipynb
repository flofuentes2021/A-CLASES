```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y1 = np.genfromtxt('Tiempos-sin-ver.txt')
print(len(y1), 'medidas')
y2 = np.genfromtxt('Tiempos-viendo.txt')
print(len(y2), 'medidas')
```

```python
p1 = np.mean(y1)
#mean define el promedio de ciertos datos
s1 = np.std(y1, ddof=1)
#std define la desviación estándar
print('Promedio sin ver =', p1)
print('Desviación estandar sin ver = ',s1)
p2 = np.mean(y2)
s2 = np.std(y2, ddof=1)
print('Promedio viendo =', p2)
print('Desviación estandar viendo = ', s2)
```

```python
plt.scatter(range(1, len(y1)+1), y1, color='blue', label = 'sin ver')
#rango de x es de 1 hasta la longitud de y1
plt.scatter(range(1, len(y2)+1), y2, color='green', label='viendo')
plt.hlines(p1,1,46, color='blue')
#hlines es una linea que traza
plt.hlines(p2,1 ,46, color='green')
plt.grid(True)
plt.xlabel('Número de medición')
plt.ylabel('Tiempo registrado [s]')
plt.legend()
#muestra la simbología
plt.xlim(0,47)
```

```python
plt.hist(y1, bins=15, range=(0,1.5), color='blue', label='sin ver', alpha=0.3)
#bins es la cantidad de barras, rango es desde donde hasta donde será el gráfico
plt.hist(y2, bins=15, range=(0,1.5), color='green', label='viendo', alpha=0.3, hatch='/')
plt.grid(True)
plt.legend()
plt.xlabel('Tiempo registrado [s]')
plt.ylabel('Número de repeticiones')
```
