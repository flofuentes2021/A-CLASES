```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos = np.genfromtxt('ejemplo-s11b.txt')
x = datos[:,0]
y = datos[:,1]
```

```python
#Gráfico de los datos
plt.scatter(x,y)
plt.grid(True)
```

```python
#Se define un ajuste polinomial de una recta
a1, a0 = np.polyfit(x,y,1)
print('a0 = ', a0)
print('a1 = ', a1)
```

```python
#Se definen los residuos
#RESIDUOS = Y - FUNCION AJUSTE
res = (y-a0-a1*x) 
st = np.sum((y - np.mean(y))**2)
sr = np.sum(res**2)
r2 = (st-sr)/st
print(r2) #mientras r2 más cercano 1 es mejor la aproximación
```

```python
#Gráfico del ajuste
xx = np.linspace(0,10,100)
plt.plot(x,a0+a1*x, color= 'red')
plt.scatter(x,y)
plt.grid(True)
#plt.xlim(0,1)
#plt.ylim(0,1)
plt.xlabel('$x$')
plt.ylabel('$y$')
```
