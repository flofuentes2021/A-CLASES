```python
import numpy as np
import matplotlib.pyplot as plt
```

# Residuos
__*Después*__ de realizar el ajuste:
\begin{equation}
\epsilon_i = y_i - f(x_i)
\end{equation}
donde f(x) es la función del modelo con los parámetros ya ajustados.

Coeficiente de determinación
$$
r^2 = (S_t - S_r)/ S_t, \qquad S_t = \sum_{i=1}^N(y_i - \bar{y})^2, \qquad S_r = \sum_{i=1}^N(y_i - f(x_i))^2 = \sum_{i=1}^N \epsilon_i^2
$$

# Ejemplo 1

```python
datos = np.genfromtxt('ejemplo-s11a.txt')
x = datos[:,0]
y = datos[:,1]
```

```python
plt.scatter(x,y)
plt.grid(True)
```

\begin{equation}
y = f(x) = a_1 x + a_0
\end{equation}

```python
a1, a0 = np.polyfit(x,y,1)
print('a0 = ', a0)
print('a1 = ', a1)
```

```python
res = (y-a0-a1*x)
st = np.sum((y - np.mean(y))**2)
sr = np.sum(res**2)
r2 = (st-sr)/st
print(r2)
```

```python
xx = np.linspace(0,10,100)
plt.plot(x,a0+a1*x, color= 'red')
plt.scatter(x,y)
plt.grid(True)
#plt.xlim(0,1)
#plt.ylim(0,1)
plt.xlabel('$x$')
plt.ylabel('$y$')
```

```python
plt.scatter(x,res)
plt.grid(True)
plt.xlabel('$x$')
plt.ylabel('$\\epsilon_i = y_i - f(x_i)$')
```

```python
xx = np.linspace(0,10,100)
plt.plot(x,a0+a1*x, color= 'red')
plt.scatter(x,y)
plt.grid(True)
plt.xlim(0,1)
plt.ylim(0,1)
plt.xlabel('$x$')
plt.ylabel('$y$')
```
