```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos = np.genfromtxt('Histograma.dat')
```

```python
T = datos[:,1]
```

```python
n, ext = np.histogram(T,bins=5)
#n es el valor de los conteos que es el primer arreglo 
#ext es el valor de los extremos
```

```python
ext[:-1]
#elimina el ultimo dato
```

```python
plt.bar(ext[:-1],n)
#Hace un gráfico de barras
```

```python
plt.style.use('fivethirtyeight')
plt.hist(T,bins=5, range={97,102},rwidth=0.98)
#genera un histograma, con 5 intervalos, con datos entre el rango
#el rwidth 98% del ancho del intervalo, más pequeño el espacio entre barras
plt.xlabel('$T$ °C')
plt.ylabel('Número de eventos')
plt.grid(True)
```

```python
plt.style.use('default')
plt.hist(T,bins=np.arange(97,103,1),rwidth=0.98)
plt.xlabel('$T$ °C')
plt.ylabel('Número de eventos')
plt.grid(True)
plt.show()
```

```python
plt.style.use('default')
plt.hist(T,bins=50,range=(97,102),rwidth=0.98)
plt.xlabel('$T$ °C')
plt.ylabel('Número de eventos')
plt.grid(True)
```

```python
N=100
n = np.random.random(size=N)
#genera numeros aleatorios
plt.hist(n,bins=10, range=(0,1), rwidth=0.98)
plt.xlabel('Valor de la variable aleatoria')
plt.ylabel('Numero de repeticiones')
plt.grid(True)
```

```python
N=10000
n = np.random.randn(N)
#genera numeros aleatorios
plt.hist(n,bins=100, rwidth=0.98)
plt.xlabel('Valor de la variable aleatoria')
plt.ylabel('Numero de repeticiones')
plt.grid(True)
```
