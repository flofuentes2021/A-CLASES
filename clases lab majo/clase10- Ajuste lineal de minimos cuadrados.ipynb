```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
datos = np.genfromtxt('Ajuste minimos cuadrados.dat')
x = datos[:,0]
y = datos[:,1]
```

```python
plt.scatter(x,y)
plt.grid(True)
plt.xlabel('$x$')
plt.ylabel('$x$')
xx = np.linspace(5,15,100)
yy = 1*xx
plt.plot(xx,yy)
```

```python
def g(a0,a1):
    plt.scatter(x,y)
    plt.grid(True)
    plt.xlabel('$x$')
    plt.ylabel('$x$')
    xx = np.linspace(5,15,100)
    yy = a0+a1*xx
    plt.plot(xx,yy)
    res = y - (a0+a1*x)
    #residuos
    chi2 = np.sum(res**2)
    #suma del cuadrado de los residuos
    print(chi2)
```

```python
g(0,1)
```
