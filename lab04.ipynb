```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#Ejercicio b) condiciones inciales
t = np.arange(0, 40, 0.1)
h = 0.1 

pi = np.zeros(len(t))
delta = np.zeros(len(t))

#Condiciones inciales 
mu = 1
pi[0] = 0.1
delta[0] = 5.0

for n in range(len(t)-1):
    pi[n+1] = pi[n] + h*pi[n]*(1-delta[n])
    delta[n+1] = delta[n] + h*mu*delta[n]*(pi[n+1]-1)
    
plt.plot(t, pi , color= "pink")
plt.plot(t, delta, color= "purple")
plt.savefig("presa-depre.pdf")
```
