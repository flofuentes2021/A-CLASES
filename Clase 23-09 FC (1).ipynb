# Solución numérica de ecuaciones diferenciales ordinarias

## Repaso derivadas númericas

Derivada adelantada -> $ f'(x)= \frac{f(x+h)-f(x)}{h}+ O(h).$
Derivada centrada de orden 2 -> $ f'(x)= \frac{f(x+h)-f(x-h)}{2h}+ O(h^2).$

SIEMPRE SE PUEDE ESCRIBIR LA EDO COMO $\frac{dx}{dt}= f(x,t) \tag{1}$ 
derivada adelantada $ \frac{dx}{dt}= \frac{f(x+h)-f(x)}{h} \tag{2}$
reemplazando (2) en (1) $x(t+h) \simeq x(t) + hf (x(t),t) \tag{3}$
 a tiempos discretos serie de tiempo $x_{n+1}= x_n + hf(x_n,t_n)$*
 

$\ddot{x} + w^2 x = -g \tag{1}$
$\dot{x}= u  \tag{2}$
$\dot{u}=\ddot{x}= -g- w^2 x \tag{3}$
$ y =  
\begin{pmatrix}
  x\\ 
 u
\end{pmatrix}$, $f(y) = 
\begin{pmatrix}
  \dot{x}\\ 
 \dot{u}
\end{pmatrix}\tag*{}$ =
$
\begin{pmatrix}
  u \\ 
 {-g- w^2 x}
\end{pmatrix}\tag{4}$.

```python
import numpy as np
import matplotlib.pyplot as plt

g=9.8
w=1.0

h=0.001
t=np.arange(0, 10.0, h)
x=np.zeros(t.size)
u=np.zeros(t.size)


x[0] =0.0
u[0] =0.0

for n in range(t.size-1):
    x[n+1]= x[n] + h*u[n] 
    u[n+1]= u[n] + h*(-g-(w**2)* x[n+1])

plt.plot(t,x)
plt.plot(t,u)
```
