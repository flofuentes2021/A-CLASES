## Método Del Salto de rana 
### (leap-frog)



es exacto para funciones del tipo $ \ddot{x}=f(X)$ -> tiene menor orden de error

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x,t, w=1.0):
    return  -w**2 *x

tmax=30.0
h=0.01
t=np.arange(0,tmax,h)

x= np.empty(t.size)
u= np.empty(t.size)

#condiciones iniciales
x[0]=0.0
v0=1.0

u[0]=v0- 0.5*h* f(x[0],t[0])

for n in range(t.size-1):
    u[n+1] = u[n]+h*f(x=x[n],t=t[n], w=3.0)
    x[n+1] = x[n]+h*u[n+1]

plt.plot(t, x , color="pink")
plt.show()
```

# Método de Runge - Kutta de 4to orden

```python
def f(u,t, w=1.0):
    x,v = u
    return  np.array([v, - w**2 *x])

tmax=30.0
h=0.01
t=np.arange(0,tmax,h)

u= np.empty([t.size,2])

#condiciones iniciales
u[0]=[0.0, 1.0]

for n in range(t.size-1):
    K1= f(u[n],t[n])
    K2= f(u[n]+0.5*h*K1,t[n]+0.5*h)
    K3= f(u[n]+0.5*h*K2,t[n]+0.5*h)
    K4= f(u[n]+h*K3,t[n]+h)
    u[n+1] = u[n] + (h/6)*(K1 + 2*K2 + 2*K3 + K4)


plt.plot(t, u[:,0] , color="pink") #grafica x
plt.plot(t, u[:,1] , color="purple") #gráfica y
plt.show()  
```
