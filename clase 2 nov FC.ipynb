```python
import numpy as np
import matplotlib.pyplot as plt
from PolyInterpolation import Lagrange
from simple_cspline import cspline
```

```python
xi = np.linspace(-4,4,9)
yi = np.exp(-xi**2)
```

```python
f = Lagrange(xi,yi)
g = cspline(xi,yi)
```

```python
plt.scatter(xi,yi, color="m", marker="*")
x=np.linspace(-4,4-1e-9,200)
plt.plot(x,f(x) , color="c")
plt.plot(x,g(x), color ="purple")
```
